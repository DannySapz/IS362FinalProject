```python
#import pandas and numpy
import pandas as pd
import numpy as np
```

1) Read in the data source and place it into dataframes

```python
url_anime =  "C:\\Users\\danie\\Downloads\\AnimeData\\anime.csv"
url_rating = "C:\\Users\\danie\\Downloads\\AnimeData\\rating.csv"
```

```python
anime = pd.DataFrame(pd.read_csv(url_anime))
rating = pd.DataFrame(pd.read_csv(url_rating))
```

```python
anime
```

```python
rating
```

2) Merge the two dataframes

```python
merged = anime.merge(rating, on='anime_id', how='left')
merged = merged.sort_values(by='anime_id', ascending=True)
merged
```

3) Find the average of the user's ratings

```python
avg = merged.reset_index().groupby("anime_id").aggregate(np.mean)
avg
```

4) Checking to make sure the values are correct and it wasn't my merge and sort that made funky values

```python
locd = merged.loc[merged['anime_id'] == 1]
(locd['rating_y'].sum())/locd['rating_y'].count()
```

```python
locd = anime.loc[anime['anime_id'] == 1]
locd
```

5) Joining the average of the users rating dataframe to the anime dataframe to have full data on ratings, genre, names, etc

```python
joined = anime.join(avg, on='anime_id', how='left', lsuffix='_anime', rsuffix='_avg', sort=True).reset_index()
joined
```

6) Drop columns that I won't be using

```python
cols_of_interest = ['level_0', 'members_anime', 'index', 'members_avg', 'user_id', 'rating_x']
df = joined[cols_of_interest]
joined.drop(df, axis=1, inplace=True)
joined
```
